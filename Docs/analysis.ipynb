```python
import pandas as pd
```

```python
cols = pd.read_csv('nas-columns.csv')
cols.head()
```

```python
labels = pd.read_csv('nas-labels.csv')
labels.head()
```

```python
pupil = pd.read_csv('nas-pupil-marks.csv')
pupil.head(10)
```

```python
data = pupil.copy()
```

```python
data.iloc[:,60:] = data.iloc[:,60:].fillna(value=0)
data.head()
```

```python
data.fillna(value='No',inplace=True)
```

```python
total = list()
for i,row in data.iterrows():
    total.append(row[60:].tolist())
```

```python
average = list()
for i,val in enumerate(total):
    count = 0
    for v in val:
        if v != 0:
            count += 1
    if count == 0:
        count = 1
    average.append(sum(val)/count)
```

```python
for i,row in data.iterrows():
    if row['Maths %'] == 0:
        data.set_value(i,"Maths %",average[i])
```

```python
for i,row in data.iterrows():
    if row['Reading %'] == 0:
        data.set_value(i,"Reading %",average[i])
```

```python
for i,row in data.iterrows():
    if row['Science %'] == 0:
        data.set_value(i,"Science %",average[i])
```

```python
for i,row in data.iterrows():
    if row['Social %'] == 0:
        data.set_value(i,"Social %",average[i])
```

```python
data.head()
```

```python
m = data['Maths %'].mean()
data['Maths %'].replace(0,m,inplace=True)
```

```python
n = data['Reading %'].mean()
data['Reading %'].replace(0,n,inplace=True)
```

```python
o = data['Science %'].mean()
data['Science %'].replace(0,o,inplace=True)
```

```python
p = data['Social %'].mean()
data['Social %'].replace(0,p,inplace=True)
```

```python
states = labels[['Level','Rename']].iloc[208:,:]
states.set_index('Level')
states.head()
```

```python
state_code = states['Level'].tolist()
state_name = states['Rename'].tolist()
```

```python
state_dict = dict()
for i,val in enumerate(state_code):
    state_dict[val] = state_name[i]
state_dict
```

```python
data.drop('STUID',axis=1,inplace=True)
```

```python
data['State'].replace(state_dict,inplace=True)
```

```python
data_2 = data[data['Gender'] != 0]
data_2['Gender'].replace([1,2],['Boy','Girl'],inplace=True)
data_for_2 = data_2[['State','Gender','Maths %','Reading %','Science %','Social %']]
data_q2_b = data_for_2[data_for_2['Gender'] == 'Boy']
```

```python
q2datab = data_q2_b.groupby(['State'],axis=0).mean()
q2datab.to_json('data3_b.json')
data_q2_g = data_for_2[data_for_2['Gender'] == 'Girl']
q2datag = data_q2_g.groupby(['State'],axis=0).mean()
q2datab.to_json('data3_g.json')
```

```python
data_for_3 = data[['State','Maths %','Science %']]
data_q3 = data_for_3.groupby(['State']).mean()
data_q3.to_json('data3.json')
```
